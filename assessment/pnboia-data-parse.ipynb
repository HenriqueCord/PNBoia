## Creating full dataframe

```python
def read_csv_files_in_directory(root_directory):
    dataframes = []

    for root, _, files in walk(root_directory):
        for file in files:
            if file.endswith('.csv'):
                file_path = path.join(root, file)
                df = pd.read_csv(file_path)
                df['origin_filename'] = f'{file}'
                dataframes.append(df)

    return dataframes
```

```python
dfs_to_concat = read_csv_files_in_directory(root_directory=DATASET_PATH)
df_buoys = pd.concat(dfs_to_concat)
```

## Trabalhando com uma slice desse dataframe

```python
def convert_column_to_datetime(df, column_name, date_format):
    """
    Convert a specified column in a DataFrame to datetime using the provided schema.

    Args:
        df (pd.DataFrame): The DataFrame containing the data.
        column_name (str): The name of the column to convert.
        distro_schema (namedtuple): The schema object containing date format information.

    Returns:
        pd.DataFrame: The DataFrame with the specified column converted to datetime.
    """
    try:
        # Convert the specified column to datetime
        df[column_name] = pd.to_datetime(
            df[column_name], format=date_format
        )
        return df
    except KeyError:
        print(f"Error: Column '{column_name}' not found in the DataFrame.")
        return None
```

## Plotting

```python
df_wind_plot = df_windspeed[
    (df_windspeed['wspd1'] >= 0) &
    (df_windspeed['flag_wspd1'] == 0) &
    (df_windspeed['flag_wdir1'] == 0) &
    (df_windspeed['origin_filename'] == 'spotter_alcatrazes_qualificados_publicos.csv')
]
```

```python
df_wind_plot.shape
```

```python
fig = px.scatter_polar(
    df_wind_plot.sample(frac=0.5),
    r='wspd1',
    theta='wdir1',
    title="Wind Speed and Direction",
    template="plotly",
    width=900,
    height=600
)

# Show the plot
fig.show()
```

```python
df_temp_plot = df_windspeed[
    (df_windspeed['sst'] >= 0) &
    (df_windspeed['origin_filename'] == 'spotter_alcatrazes_qualificados_publicos.csv')
]
```

```python
fig_2 = px.line(
    data_frame=df_temp_plot,
    y='sst',
    title="Sea Surface Temperature over Time",
    width=900,
    height=600
)

fig_2.show()
```
